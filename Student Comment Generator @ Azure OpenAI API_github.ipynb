```python
# Batch import at once (10 per call)

import json    
import pandas as pd    
import openai  
import time  
  
# Set OpenAI configuration settings  
openai.api_type = "azure"  
openai.api_base = "your azure openai endpoint url"  
openai.api_version = "2023-03-15-preview"  
openai.api_key = "Azure OpenAI API Key"  
  
# Read the Excel file into a DataFrame  
df = pd.read_excel("comment_import.xlsx")  
  
# Convert the DataFrame to a list of dictionaries  
data = [row.to_dict() for _, row in df.iterrows()]  
  
# Split the data into batches of 10  
batch_size = 10  
batches = [data[i:i+batch_size] for i in range(0, len(data), batch_size)]  
  
# Loop through the batches and send requests  
for i, batch in enumerate(batches):  
    print(f"Sending request for batch {i+1}/{len(batches)} to Azure OpenAI endpoint...\n")  
  
    # Convert the batch of data to JSON  
    json_data = {"data": batch}  
    json_text = json.dumps(json_data, ensure_ascii=False)  
  
    # Send request to Azure OpenAI model  
    response = openai.ChatCompletion.create(  
         engine="Your Deploy name",  
         temperature=0.8,  
         max_tokens=1500,  
         top_p=0.95,  
         frequency_penalty=0,  
         presence_penalty=0,  
         messages=[  
             {"role": "system", "content": "扮演品德教育科老師，以第三人稱為每一位學生分別寫一段評語"},  
             {"role": "user", "content": json_text}  
         ]  
     )  
  
    # Print the response from the Azure OpenAI model  
    print(f"Azure OpenAI Return for batch {i+1}/{len(batches)}: {response.choices[0].message.content}\n")  
  
    # Wait for 1 second before sending the next request  
    time.sleep(1)  
```
